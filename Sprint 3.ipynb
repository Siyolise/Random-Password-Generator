```python
#This is a Password generator program
#Every time you run the program, it outputs a random password.
import random

def myPassword(pswLength=12):
    '''This is a default function that returns a random password
    in length of 12 different characters such as lowerCaseLetters, 
    upperCaseLetters,numbers and symbols'''
    
    #prints introduction to the program and produces a newline after.
    print("This is a Random Password Generator","\n")
    
    #store (alphabet,numbers and symbols) characters in a variable named password_chars.
    #From random module,use sample() function for non-repetitive selection of password- 
    #characters in length of 12 and store it in a variable named password_gen.
    #Finally,return a password generated by the program without comma-seperated values created by default.
    password_chars = "abcdefghijklmnopqrsvwxyzABCDEFGHIJKLMNOPQRSVWXYZ1234567890!@#$%*?+&"
    password_gen = random.sample(password_chars,pswLength) 
    return "".join(password_gen)
    

print("Your secret password is =", myPassword()) 
print("Please do not share it with anyone.")
```
